```r
library(dplyr)
Data <- read.csv(file="data3.csv", header=TRUE, sep=",")
```

```r
new <- filter(Data,eminuse==1 | intmob==1,home4nw==1) %>%
    select(age,inc,educ2,bbhome1,intfreq)
new
```

### I choose five columns of data: age inc educ2 bbhome1 and intfreq. Now the way of normalizing my data is as following: 
#### 1. For educ2: 
1	Less than high school (Grades 1-8 or no formal schooling) -> 10  

2	High school incomplete (Grades 9-11 or Grade 12 with NO diploma) ->20  

3	High school graduate (Grade 12 with diploma or GED certificate)  ->30  

4	Some college, no degree (includes some community college)   ->40 

5	Two year associate degree from a college or university   ->50  

6	Four year college or university degree/Bachelor’s degree (e.g., BS, BA, AB) ->60  

7	Some postgraduate or professional schooling, no postgraduate degree -> 70  

8	Postgraduate or professional degree, including master’s, doctorate, medical or law degree (e.g., MA, MS, PhD, MD, JD) -> 80  

98	Don’t know ->0  

99	Refused    ->0  

#### 2. For bbhome1:
1	Dial-up -> 10 

2	Higher-speed ->100

3	(VOL.) Both Slow-speed/Dial-up and Higher-speed/Broadband ->60

4	(VOL.) Access internet only using cell phone or tablet   ->40

5	(VOL.) No home internet access ->0

8	(VOL.) Don't know  ->0

9	(VOL.) Refused   ->0


#### 3. For intfreq:

1	Almost constantly ->100

2	Several times a day ->80

3	About once a day  ->60

4	Several times a week, OR -> 40

5	Less often? -> 20

8	(VOL.) Don’t know -> 0

9	(VOL.) Refused -> 0

#### 4. For inc:
1	Less than $10,000 -> 5000

2	10 to under $20,000 -> 15000

3	20 to under $30,000 -> 25000

4	30 to under $40,000 -> 35000

5	40 to under $50,000 -> 45000

6	50 to under $75,000 -> 60000

7	75 to under $100,000 -> 90000

8	100 to under $150,000 -> 125000

9	$150,000 or more -> 150000

98	(VOL.) Don't know -> 0

99	(VOL.) Refused  -> 0






```r
new$inc[new$inc==1] <- 5000
new$inc[new$inc==2] <- 15000
new$inc[new$inc==3] <- 25000
new$inc[new$inc==4] <- 35000
new$inc[new$inc==5] <- 45000
new$inc[new$inc==6] <- 60000
new$inc[new$inc==7] <- 90000
new$inc[new$inc==8] <- 125000
new$inc[new$inc==9] <- 150000
new$inc[new$inc==98] <- 0
new$inc[new$inc==99] <- 0
```

```r
new$intfreq[new$intfreq==1] <- 100
new$intfreq[new$intfreq==2] <- 80
new$intfreq[new$intfreq==3] <- 60
new$intfreq[new$intfreq==4] <- 40
new$intfreq[new$intfreq==5] <- 20
new$intfreq[new$intfreq==8] <- 0
new$intfreq[new$intfreq==9] <- 0
```

```r
new$bbhome1[new$bbhome1==1] <- 10
new$bbhome1[new$bbhome1==2] <- 100
new$bbhome1[new$bbhome1==3] <- 60
new$bbhome1[new$bbhome1==4] <- 40
new$bbhome1[new$bbhome1==5] <- 0
new$bbhome1[new$bbhome1==8] <- 0
new$bbhome1[new$bbhome1==9] <- 0
```

```r
new$educ2[new$educ2==1] <- 10
new$educ2[new$educ2==2] <- 20
new$educ2[new$educ2==3] <- 30
new$educ2[new$educ2==4] <- 40
new$educ2[new$educ2==5] <- 50
new$educ2[new$educ2==6] <- 60
new$educ2[new$educ2==7] <- 70
new$educ2[new$educ2==8] <- 80
new$educ2[new$educ2==98] <- 0
new$educ2[new$educ2==99] <- 0
```

```r
new
```

#### Now use k-means for k=3

```r
k <-kmeans(new, centers=3)
k$centers
table(k$cluster)
```

#### Conclusion:  
1. The age of three clusters are almost same. 

2. Higher income, higer education level, higher internet speed at home, higer internet using frequency. 

3. Most of participants in the survey have low income.

#### Now for k=5

```r
k <-kmeans(new, centers=5)
k$centers
table(k$cluster)
```

#### Conclusion:
1. Almost remains the same as it's in k=3.

2. Cluster 2, which has highest income, also has highest education level and highest internet using frequency.

3. Cluster 5, which has only 8 members, only got 1.25 for bbhome1 and lowest internet using frequency. Meanwhile, it has highest age. Maybe elders can't get used to internet?
