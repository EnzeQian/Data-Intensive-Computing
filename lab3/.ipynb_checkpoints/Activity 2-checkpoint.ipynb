```r
data<-read.csv("german_credit_data1.csv",header=T)
data
```

```r
data$Sex <- as.character(data$Sex)
data$Housing <- as.character(data$Housing)
data$Saving.accounts <- as.character(data$Saving.accounts)
data$Checking.account <- as.character(data$Checking.account)
data$Purpose  <- as.character(data$Purpose)
str(data)
data
```

```r
data$Sex[data$Sex=="male"] <- 1
data$Sex[data$Sex=="female"] <- 2

data$Housing[data$Housing=="own"] <- 1
data$Housing[data$Housing=="free"] <- 2
data$Housing[data$Housing=="rent"] <- 3

data$Saving.accounts[is.na(data$Saving.accounts)] <- 0
data$Saving.accounts[data$Saving.accounts=="little"] <- 1
data$Saving.accounts[data$Saving.accounts=="moderate"] <- 2
data$Saving.accounts[data$Saving.accounts=="quite rich"] <- 3
data$Saving.accounts[data$Saving.accounts=="rich"] <- 4

data$Checking.account[is.na(data$Checking.account)] <- 0
data$Checking.account[data$Checking.account=="little"] <- 1
data$Checking.account[data$Checking.account=="moderate"] <- 2
data$Checking.account[data$Checking.account=="quite rich"] <- 3
data$Checking.account[data$Checking.account=="rich"] <- 4

data$Purpose[data$Purpose=="radio/TV"] <-1
data$Purpose[data$Purpose=="education"] <-2
data$Purpose[data$Purpose=="furniture/equipment"] <-3
data$Purpose[data$Purpose=="car"] <-4
data$Purpose[data$Purpose=="business"] <-5
data$Purpose[data$Purpose=="repairs"] <-6
data$Purpose[data$Purpose=="domestic appliances"] <-7
data$Purpose[data$Purpose=="vacation/others"] <-8
data
```

```r
n.points<-1000
sampling.rate <-0.8
num.test.set.labels<- n.points* (1 - sampling.rate)

training<-sample(1:n.points, sampling.rate*n.points, replace=FALSE)
train<-subset(data[training,], select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Duration,Purpose,Credit.amount))

testing<-setdiff(1:n.points,training)
test<-subset(data[testing,], select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Duration,Purpose,Credit.amount))
```

```r
library(class)
```

```r
cl<- data$Credit.Risks[training]
true.labels<-data$Credit.Risks[testing]
```

```r
result <- matrix(c("k","rate"),ncol=2,byrow=TRUE)
```

```r
for(k in 1:20) {
  predicted.labels <- knn(train,test,cl,k)
  num.incorrect.labels<-sum(predicted.labels !=true.labels)
  misclassification.rate<-num.incorrect.labels/num.test.set.labels
  result <- rbind(result, c(k, misclassification.rate))
}
result
```

```r
sampling.rate <-0.7
num.test.set.labels<- n.points* (1 - sampling.rate)

training<-sample(1:n.points, sampling.rate*n.points, replace=FALSE)
train<-subset(data[training,], select=c(Age,Credit.amount))

testing<-setdiff(1:n.points,training)
test<-subset(data[testing,], select=c(Age,Credit.amount))

cl<- data$Credit.Risks[training]
true.labels<-data$Credit.Risks[testing]

result <- matrix(c("k","rate"),ncol=2,byrow=TRUE)

for(k in 1:20) {
  predicted.labels <- knn(train,test,cl,k)
  num.incorrect.labels<-sum(predicted.labels !=true.labels)
  misclassification.rate<-num.incorrect.labels/num.test.set.labels
  result <- rbind(result, c(k, misclassification.rate))
}
result
```
